```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as gd
```

## Scrapping Website

```python
url = "https://www.flipkart.com/laptops-store?s_kwcid=AL!739!3!603380124709!e!!g!!flipkart&gclsrc=aw.ds&&semcmpid=sem_8024046704_sitelink_laptop_goog&gclid=EAIaIQobChMI25uA7NC0-AIVCJlmAh3SmwPfEAAYASADEgJdPvD_BwE"
```

## Website Response

```python
res = requests.get(url)
```

```python
res
```

## Parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

## Laptops Name

```python
laptops = soup.select(".s1Q9rs")
```

```python
laptops
```

```python
prices = soup.select('._30jeq3')
```

```python
prices
```

```python
price_title=[]
```

```python
for item in prices:
    price_title.append(item.text)
```

```python
price_title
```

```python
laptops_names=[]
```

```python
for item in laptops:
    laptops_names.append(item.text)
```

```python
laptops_names
```

## Data Frame

```python
df=gd.DataFrame(laptops_names,columns=['laptops names'])
```

```python
df
```

```python
df['price_title']=price_title
```

```python
df
```

```python
df.to_csv('sg.csv')
```
